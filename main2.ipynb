```python
import pandas as pd
import ast
```

```python
df_reviews = pd.read_csv('datasets/reviews.csv')
```

```python
list_users = df_reviews.user_id.values
```

```python
list_reviews = []
for i in df_reviews.reviews:
    list_reviews.append(ast.literal_eval(i))
```

```python
list_reviews[0].keys()
```

```python
list_tuples = []

for i, r in enumerate(list_reviews):
    r['id_user'] = list_users[i]
    k, v = r.keys(), r.values()
    list_tuples.append(tuple(v))

df_reviews_clean = pd.DataFrame(list_tuples, columns=['funny', 'posted', 'last_edited', 'item_id', 'helpful', 'recommend', 'review', 'id_user'])
```

```python
df_reviews_clean.to_csv('reviews_clean.csv')
```

```python
funny = []
for i in df_reviews_clean.funny.values:
    temp = str(i).split() 
    if len(temp) == 0:
        funny.append(0)
    else:
        funny.append(int(temp[0].replace(',','')))

print(funny)
```

```python
df_reviews_clean['funny'] = pd.Series(funny)

fechas_posted_reviews = df_reviews_clean.iloc[:,1].apply(lambda x: pd.to_datetime(x[7:-1]) if ',' in x else pd.to_datetime(x[7:-1] + ', 2016'))

fechas_posted_reviews
```

```python
fechas_posted_reviews.to_csv('fechas_posted_reviews.csv')
```
